```python
page = requests.get("https://www.yucatan.com.mx/")
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
panel= soup.find(class_="wpb_column vc_column_container vc_col-sm-8")
```

```python
forecast_items = panel(class_="entry-body")
```

```python
tonight = forecast_items[0]
```

```python
print(tonight.prettify())
```

```python
period = tonight.find(class_="entry-header").get_text()  
```

```python
print(period)
```

```python
period_tags = panel.select(".entry-body .entry-header")
```

```python
periods = [period_tags.get_text() for period_tags in period_tags]
```

```python
periods.rstrip()
```

```python
import pandas as pd
```

```python
weather = pd.DataFrame({
    "noticias": periods,
   
})
weather
```
