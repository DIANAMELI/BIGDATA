```python
import requests
```

```python
page = requests.get("https://www.informador.mx/")
```

```python
page
```

```python
pagess=(page.content.strip())
```

```python
print(pagess)
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(pagess, 'html.parser')
```

```python
print(soup.prettify())
```

```python
panel= soup.find(class_="col-main")
```

```python
forecast_items = panel(class_="news-data")
```

```python
tonight = forecast_items[0]
```

```python
print(tonight.prettify())
```

```python
period = tonight.find(class_="news-title").get_text()  
```

```python
print(period)
```

```python
period_tags = panel.select(".news-data .news-title")
```

```python
periods = [pt.get_text() for pt in period_tags]
```

```python
periods
```

```python
import pandas as pd
```

```python
weather = pd.DataFrame({
    "noticias": periods,
   
})
weather
```
