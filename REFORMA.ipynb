```python
page = requests.get("https://www.reforma.com/")
```

```python
page.content
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
panel= soup.find(id="zonanoticias")
```

```python
forecast_items = panel(class_="caption")
   
```

```python
tonight = forecast_items[0]
```

```python
print(tonight.prettify())
```

```python
period = tonight.find(class_="ic_caption").get_text()  
```

```python
print(period)
```

```python
period_tags = panel.select(".caption .ic_caption")
```

```python
periods = [pt.get_text() for pt in period_tags]
```

```python
periods 
```

```python
import pandas as pd
```

```python
weather = pd.DataFrame({
    "noticias": periods,
   
})
weather
```
