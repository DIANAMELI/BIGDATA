# --ANALIZAR CON BEAUTIFULSOUP--

# A√ëADIR LA BIBLIOTECA

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')
```

# IMPRIMIR EL TEXTO ANALIZADO 

```python
print(soup.prettify())
```

# EXTRAER LO QUE SE ENCUENTRA DENTRO DE LAS CLASES

```python
panel= soup.find(class_="panel-panel grid-11 alpha")
```

```python
forecast_items = panel(class_="multimedia")##
```

```python
tonight = forecast_items[0]
```

```python
print(tonight.prettify())
```

```python
period = tonight.find(class_="field-content").get_text()
```

```python
print(period)
```

# EXTRAER LOS ELEMENTOS DE UNA CLASE

```python
period_tags = panel.select(".multimedia .field-content ")
```

```python
periods = [pt.get_text() for pt in period_tags]
```

```python
periods 
```

# --USAR DATAFRAME-- (Analizar los datos)

# IMPORTAR PANDAS

```python
import pandas as pd
```
